```python
import pandas as pd
import numpy as np
from skimpy import skim
from sklearn.preprocessing import MinMaxScaler

test = pd.read_csv('test.csv')
test_calendar = pd.read_csv('test_calendar.csv')
train = pd.read_csv('train.csv')
train_calendar = pd.read_csv('train_calendar.csv')

train = train.drop(['holiday_name', 'id'], axis=1)
test = test.drop(['holiday_name','id'], axis=1)

for col in train.columns:
    if train[col].dtype != 'int64':
        train[col] = pd.factorize(train[col])[0]

for col in test.columns:
    if test[col].dtype != 'int64':
        test[col] = pd.factorize(test[col])[0]

scaler = MinMaxScaler(feature_range=(1, 500))
scaler.fit(train)
train = pd.DataFrame(scaler.transform(train))

# for column in test:
#     print(test[column].value_counts())

from sklearn.model_selection import train_test_split
from sklearn import tree

#split data
# X_train,y_train = train.drop(['orders'], axis=1), train['orders']
# X_test,y_test = test.drop(['orders'], axis=1), test['orders']
X,y = train.drop([2], axis=1), train[2]
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.70)

#train data
trialData = {'numNodes':[],'mape':[]}

#add loop for features
#for k in range(3,34,3):
for i in range(2,15):
    #why is MSE inconsistent when start range from 5?
    #why is MSE very regular when start from 2?
    clf = tree.DecisionTreeRegressor(max_leaf_nodes = i)
    clf = clf.fit(X_train,y_train)
    #estimate quality of prediction
    prediction = clf.predict(X_test)

    #mean squared error (MSE) --> estimate accuracy
    error = 0
    for j in range(len(y_test)):
        error += abs((prediction[j] - y_test.iloc[j])/y_test.iloc[j])
    error = 100*(error/len(y_test))

    trialData['numNodes'].append(i)
    trialData['mape'].append(error)

trialData = pd.DataFrame(trialData)
display(trialData)

#display(train_calendar)
display(train)
display(test)
skim(test)
```
